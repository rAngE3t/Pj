```python
def eda_txt(data,name):
    lower=" ".join(data[name]).islower()
    html=data[name].apply(lambda x:True if re.search("<.+>",x) else False).sum()
    url=data[name].apply (lambda x:True if re.search("http[s]?://.+? +",x) else False).sum()
    char=data[name].apply(lambda x:True if re.search("[].$*[#@!\":;%_0-9]",x)else False).sum()
    if lower==False:
        print("not lower")
    if html>0:
        print("html tag present")
    if url>0:
        print("url present")
    if char>0:
        print("characters present")
```
